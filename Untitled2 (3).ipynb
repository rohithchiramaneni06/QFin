```python
def fetch_data(tickers, start='2010-01-01', end=None, use_synthetic=False, seed=42):
  print("Downloading price data from Yahoo Finance...")
  data = yf.download(tickers, start=start, end=end)['Close']
  data = data.interpolate(method='time', limit_direction='both', axis=0)
  returns = data.pct_change().dropna()
  return returns
```

```python

def compute_mu_cov(prices):
  mu = prices.mean(axis=0)* 252.0
  cov = prices.cov() * 252.0
  return mu, cov


def train_and_save_stats(prices, out_path='portfolio_stats.joblib'):
  mu, cov = compute_mu_cov(prices)

  tickers = prices.columns.tolist()
  joblib.dump({'mu': mu, 'cov': cov, 'tickers': tickers}, out_path)
  print(f'Saved stats to {out_path} (n_assets={len(tickers)})')
  return mu, cov, tickers
```

```python
from qiskit_optimization import QuadraticProgram
from qiskit_optimization.converters import QuadraticProgramToQubo

def build_qubo(mu, cov_matrix, user_risk, k, assets):
    """
    Build a QUBO for portfolio optimization with adaptive risk scaling.

    Objective: minimize (-mu^T x + λ x^T cov x)
    where λ depends on user risk preference and market statistics.

    Args:
        mu : pd.Series of expected returns (annualized).
        cov_matrix : pd.DataFrame covariance matrix (annualized).
        user_risk : float in [0, 1] (higher → more risk tolerance).
        k : int, number of assets to pick.
        tickers : list of asset names (same order as mu and cov_matrix).

    Returns:
        qubo : QuadraticProgram in QUBO form (for quantum solvers).
    """
    n = len(assets)
    qp = QuadraticProgram()

    # Binary decision variable for each asset
    for t in tickers:
        qp.binary_var(name=t)

    # Compute adaptive lambda
    avg_return = mu.mean()
    avg_vol = np.sqrt(np.mean(np.diag(cov_matrix)))

    # Invert lambda: low user_risk → high penalty on variance, high user_risk → low penalty
    # Added non-linear scaling for better spread
    # Scale factor to give more weight to return when user selects higher risk
    # Map user risk (0-1) to lambda (0.1 – 10) non-linearly
    lam_min, lam_max = 0.1, 10
    lam = lam_min + (1 - user_risk)**2 * (lam_max - lam_min)

    # multiply by factor 10 to amplify impact
    print(f"Lambda scaling factor: {lam:.4f}")


    # scaling factor 10 to adjust impact

    # Objective: Minimize variance - lambda * return
    linear = {assets[i]: -mu.iloc[i] for i in range(n)}
    quadratic = {(assets[i], assets[j]): lam * cov_matrix.iloc[i, j] for i in range(n) for j in range(n)}
    avg_vol = np.sqrt(np.mean(np.diag(cov_matrix)))
    norm_factor = lam / avg_vol  # normalize return by volatility

    # Step 3: scale linear part by normalized factor
    linear_scaled = {asset: val * norm_factor for asset, val in linear.items()}

    # Step 4: scale quadratic part by lambda
    quadratic_scaled = {key: val * lam for key, val in quadratic.items()}
    qp.minimize(linear=linear_scaled, quadratic=quadratic_scaled)


    # Cardinality constraint: select exactly k assets
    qp.linear_constraint(
        linear={t: 1 for t in tickers},
        sense="==",
        rhs=k,
        name="pick_k_assets"
    )

    # Convert to QUBO
    qubo = QuadraticProgramToQubo().convert(qp)
    return qubo
```

```python
from qiskit.primitives import Sampler

def solve_qubo_with_qaoa(qubo, reps=2, maxiter=200):
    optimizer = COBYLA(maxiter=maxiter)

    # Sampler does not take a seed
    sampler = Sampler()  # exact expectation-based

    # Initialize QAOA
    qaoa = QAOA(sampler=sampler, reps=reps, optimizer=optimizer)

    # Solve with MinimumEigenOptimizer
    solver = MinimumEigenOptimizer(qaoa)
    result = solver.solve(qubo)
    return result
```

```python
def compute_portfolio_metrics(selection_vec, mu, cov, tickers, user_risk, risk_free=0.02, returns=None):
    """
    Compute fractional-weight portfolio metrics from QAOA selection vector.

    Args:
        selection_vec : list or np.array of binary values from QAOA (or continuous [0,1]).
        mu : pd.Series of expected returns (annualized).
        cov : pd.DataFrame covariance matrix (annualized).
        tickers : list of asset names (same order as mu and cov).
        user_risk : float, scales the fractional weights to match investor risk preference.
        risk_free : float, risk-free rate (for Sharpe/Sortino).
        returns : pd.DataFrame of historical returns (optional, for Sortino ratio).

    Returns:
        dict with portfolio metrics:
            - selected_assets
            - weights_fractional
            - expected_return
            - volatility
            - sharpe_ratio
            - sortino_ratio (if returns provided)
            - downside_deviation (if returns provided)
            - selection_vector
    """

    # Convert selection to binary
    selection = np.array([1 if float(x) > 0 else 0 for x in selection_vec])
    idx = np.where(selection == 1)[0].tolist()
    if len(idx) == 0:
        return None

    # Selected assets
    selected_mu = mu.iloc[idx]
    selected_cov = cov.iloc[idx, idx]

    # Preliminary weights proportional to return / variance
    weights_raw = selected_mu / np.diag(selected_cov)
    weights_fractional = weights_raw / weights_raw.sum()

    # Scale by user risk
    portfolio_var = weights_fractional.values @ selected_cov.values @ weights_fractional.values
    portfolio_vol = np.sqrt(portfolio_var)
    scaling_factor = user_risk / (portfolio_vol + 1e-12)
    weights_fractional = weights_fractional * scaling_factor

    # Normalize to sum to 1
    weights_fractional = weights_fractional / weights_fractional.sum()

    # Portfolio metrics
    exp_return = float(np.dot(weights_fractional.values, selected_mu))
    vol = float(np.sqrt(weights_fractional.values @ selected_cov.values @ weights_fractional.values))
    sharpe = float((exp_return - risk_free) / (vol + 1e-12))

    # Sortino ratio (optional)
    sortino = None
    downside_dev = None
    if returns is not None:
        selected_returns = returns.iloc[:, idx] @ weights_fractional.values
        downside_returns = np.minimum(selected_returns - risk_free, 0)
        downside_dev = np.sqrt((downside_returns**2).mean())
        if downside_dev > 0:
            sortino = float((exp_return - risk_free) / downside_dev)

    return {
        "selected_assets": [tickers[i] for i in idx],
        "weights_fractional": weights_fractional,
        "expected_return": exp_return,
        "volatility": vol,
        "sharpe_ratio": sharpe,
        "sortino_ratio": sortino,
        "downside_deviation": downside_dev,
        "selection_vector": selection
    }
```

```python
import numpy as np

def monte_carlo_portfolio(mu, cov, weights, horizon_days=30, n_sims=5000, seed=123, alpha=0.05):
    """
    Monte Carlo simulation of portfolio returns with VaR and CVaR.

    Args:
        mu : pd.Series of expected returns (annualized).
        cov : pd.DataFrame covariance matrix (annualized).
        weights : np.array or pd.Series of portfolio weights (should sum to 1).
        horizon_days : int, time horizon in trading days.
        n_sims : int, number of Monte Carlo simulations.
        seed : int, random seed for reproducibility.
        alpha : float, significance level for VaR/CVaR (default 0.05).

    Returns:
        dict with:
            - mean_return
            - std_return
            - VaR_return (at alpha)
            - CVaR_return (expected shortfall)
            - VaR_loss (positive number)
            - CVaR_loss (positive number)
            - sim_returns (all simulated portfolio returns)
    """
    np.random.seed(seed)

    # Convert horizon from daily to specified period
    h = max(1, int(horizon_days))
    mu_h = mu * (h / 252.0)
    cov_h = cov * (h / 252.0)

    # Simulate multivariate normal returns
    sims = np.random.multivariate_normal(mean=mu_h, cov=cov_h, size=n_sims)

    # Portfolio returns under each simulation
    port_rets = sims.dot(weights)

    # Statistics
    mean_sim = float(np.mean(port_rets))
    std_sim = float(np.std(port_rets))

    # VaR and CVaR
    var_ret = float(np.percentile(port_rets, 100 * alpha))  # e.g. 5% percentile
    cvar_ret = float(np.mean(port_rets[port_rets <= var_ret])) if np.any(port_rets <= var_ret) else var_ret

    # Loss equivalents (positive numbers for easier interpretation)
    var_loss = -var_ret
    cvar_loss = -cvar_ret

    return {
        "mean_return": mean_sim,
        "std_return": std_sim,
        "VaR_return": var_ret,
        "CVaR_return": cvar_ret,
        "VaR_loss": var_loss,
        "CVaR_loss": cvar_loss,
        "sim_returns": port_rets
    }
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_mc_hist(sim_returns, alpha=0.05, title='Monte Carlo Distribution of Portfolio Returns', bins=60):
    """
    Plot Monte Carlo simulated portfolio returns with VaR and CVaR lines.

    Args:
        sim_returns : np.array of simulated portfolio returns.
        alpha : float, significance level for VaR/CVaR (default 0.05).
        title : str, plot title.
        bins : int, number of histogram bins.
    """
    mean_ret = np.mean(sim_returns)
    var_ret = np.percentile(sim_returns, 100*alpha)
    cvar_ret = np.mean(sim_returns[sim_returns <= var_ret]) if np.any(sim_returns <= var_ret) else var_ret

    plt.figure(figsize=(10,5))

    # Histogram
    plt.hist(sim_returns, bins=bins, alpha=0.75, color='skyblue', edgecolor='black')

    # Mean line
    plt.axvline(mean_ret, color='green', linestyle='-', linewidth=2, label=f'Mean: {mean_ret:.4f}')

    # VaR line
    plt.axvline(var_ret, color='red', linestyle='--', linewidth=2, label=f'{int(alpha*100)}% VaR: {var_ret:.4f}')

    # CVaR line
    plt.axvline(cvar_ret, color='orange', linestyle='-.', linewidth=2, label=f'{int(alpha*100)}% CVaR: {cvar_ret:.4f}')

    # Labels and grid
    plt.title(title)
    plt.xlabel('Portfolio Return')
    plt.ylabel('Frequency')
    plt.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# SETTINGS
ASSETS = [
    "AAPL", "MSFT", "NVDA", "GOOGL", "AMZN", "META",
    "JNJ", "KO", "JPM", "BRK-B", "ORCL", "GE",     # 12 equities
    "IEF",  # US Treasuries ETF (Debt)
    "HYG",  # High Yield Corporate Bond ETF (Debt)
    "GLD", "IAU",  # Gold ETFs
    "BTC-USD"  # Bitcoin (Crypto)
]
START = '2010-01-01'
END = datetime.today().strftime('%Y-%m-%d')

# Train
prices_hist = fetch_data(ASSETS, start=START, end=END)
mu, cov, tickers = train_and_save_stats(prices_hist)


# Live window
live_prices = prices_hist.iloc[-252:]


# User inputs
K = int(input("enter no.of Stocks to be included?"))
GAMMA = int(input("enter the risk volatility?"))  # user risk
RISK_FREE = 0.02
MC_HORIZON_DAYS = 30
MC_SIMS = 5000
ALPHA = 0.05
SEED = 123
Amount = int(input("Amount to be invested?"))
Tenure = int(input("No.of Years to be invested?"))


# Optimize
qubo = build_qubo(mu, cov, GAMMA, K, tickers)
result = solve_qubo_with_qaoa(qubo, reps=2, maxiter=150)
metrics = compute_portfolio_metrics(
    selection_vec=result.x,
    mu=mu,
    cov=cov,
    tickers=list(mu.index),
    user_risk=GAMMA,
    risk_free=RISK_FREE
)



# Monte Carlo
selected_idx = [i for i, x in enumerate(result.x) if x >= 0.5]
mu_sel = mu.iloc[selected_idx]
cov_sel = cov.iloc[selected_idx, selected_idx]
weights_sel = metrics['weights_fractional'].values

mc_stats = monte_carlo_portfolio(
    mu=mu_sel,
    cov=cov_sel,
    weights=weights_sel,
    horizon_days=MC_HORIZON_DAYS,
    n_sims=MC_SIMS,
    seed=SEED,
    alpha=ALPHA
)



# Plot
plot_mc_hist(
    sim_returns=mc_stats['sim_returns'],
    alpha=ALPHA,
    title=f'{MC_HORIZON_DAYS}-day Monte Carlo Portfolio Returns'
)



# Final Report
print("Selected assets:", metrics["selected_assets"])
print("Fractional weights:\n", metrics["weights_fractional"])
print("Expected return:", metrics["expected_return"])
print("Volatility:", metrics["volatility"])
print("Sharpe ratio:", metrics["sharpe_ratio"])
print("Sortino ratio:", metrics["sortino_ratio"])

# print(f"VaR ({int(ALPHA*100)}%) return: {mc_stats['VaR_return']:.3%}")
# print(f"CVaR ({int(ALPHA*100)}%) return: {mc_stats['CVaR_return']:.3%}")
# print(f"VaR ({int(ALPHA*100)}%) loss: {mc_stats['VaR_loss']:.3%}")
# print(f"CVaR ({int(ALPHA*100)}%) loss: {mc_stats['CVaR_loss']:.3%}")

print("Monte Carlo Mean Return:", mc_stats["mean_return"])
print("Monte Carlo Std Dev:", mc_stats["std_return"])
print("VaR (5%):", mc_stats["VaR_loss"])
print("CVaR (5%):", mc_stats["CVaR_loss"])
```
